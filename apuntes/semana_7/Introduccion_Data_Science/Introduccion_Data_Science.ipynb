# Introducción al Data Science

Las herramientas con las que cuenta un Data Scientist son muy potentes y por lo tanto nos gustaría que tuviéseis algo de experiencia con esas herramientas como parte de este módulo. 

Pero lo primero que tenemos que definir es ¿qué es Data Science? 

image.png


El concepto de Data Science es el punto en el que se juntan tres habilidades cruciales:
- El conocimiento experto sobre el área en el que queremos resolver un problema
- El conocimiento matemático y de estadística para validar que los resultados que obtenemos son relevantes
- La habilidad de programación para la creación de modelos de machine learning para poder resolver ese problema


Como se puede ver en la imagen, cuando tenemos solo dos de estas habilidades hablaríamos de otros conceptos: 
- Si contamos con el conocimiento experto en el área del problema y el conocimiento matemático, estaríamos en lo que llamaríamos la investigación tradicional.
- Si contamos con el conocimiento matemático y la habilidad de programación, estaríamos en el área de machine learning
- Si contamos con la habilidad de programación y el conocimiento experto en el área del problema, estaríamos en lo que se suele llamar "la zona de peligro". Esto es así porque podemos obtener fácilmente resultados para el problema que intentamos resolver pero, sin el conocimiento matemático que nos permita comprobar la validez de dichos resultados, podemos obtener resultados poco relevantes o directamente erróneos. 



El término "machine learning" es uno de los términos más frecuentemente usados y populares hoy en día. Es muy probable que hayas escuchado este término al menos una vez si tienes algún tipo de familiaridad con la tecnología, no importa el sector en que trabajes. Aún así, las mecánicas del machine learning son un misterio para la mayoría de la gente. Para un principiante en machine learning, el tema puede parecer intimidante. Es por esto que es importante entender lo que realmente es el machine learning y aprender sobre el tema poco a poco

## Clasificación de los modelos de machine learning

Ahora que hemos definido que es el data science, cuáles son sus principales habilidades y hemos visto qué es el machine learning y cómo se diferencia de otros términos como inteligencia artificial y deep learning, podemos empezar a ver cómo se clasifican los distintos modelos de machine learning dependiendo de cómo aprenden los modelos de los datos disponibles. 

Esta clasificación la iremos repasando según vayamos viendo los distintos tipos de modelos de machine learning, pero es una forma muy rápida de ayudarnos a escoger qué tipo de modelo debemos probar a utilizar cuando nos enfrentamos a un problema. 

### Supervised learning / Aprendizaje supervisado

Los modelos de aprendizaje supervisado son aquellos en los que hay un supervisor ominisciente, el data scientist, que puede proporcionar todas las respuestas correctas en todos los escenarios de entrenamiento del modelo. Es decir, el data scientist cuenta con los valores reales que tiene que predecir el modelo de machine learning y se los proporciona al modelo en todos los escenarios de entrenamiento para que aprenda de esos datos. 

Los resultados del modelo son independientes entre sí, es decir que el resultado para unos datos de entrada es totalmente independiente de los resultados previos producidos por el modelo. 

Ejemplos de este tipo de modelos son regresión lineal, los clasificadores, las series temporales, algunos modelos de procesado de lenguaje natural, etc...

image.png

### Unsupervised learning / Aprendizaje no supervisado

Los modelos de aprendizaje no supervisado son aquellos en los que hay un supervisor, el data scientist, pero en este caso el supervisor no cuenta con los valores reales que el modelo tiene que producir y por lo tanto el modelo debe encontrar la estructura oculta en los datos sin marcar por su cuenta, explotando las diferencias, similitudes o patrones que pueda encontrar en los datos que el data scientist le proporcione. 

Al igual que en los modelos de aprendizaje supervisado, los resultados del modelo son independientes entre sí. 

Ejemplosde este tipode modelos son clustering, algunos tipos de recomendadores, 
algunos tipos de modelos de procesado de lenguaje natural,etc....

image.png

### Reinforcement learning o semi-supervised learning / Apredizaje por refuerzo o aprendizaje semi supervisado

Los modelos de aprendizaje por refuerzo (también llamados de aprendizaje semi supervisado) se diferencian de los anteriores tipos de modelos en que existe un supervisor, el data scientist, que describe el escenario en el que va a interactuar el modelo, las distintas acciones que puede ejecutar el modelo y las recompensas asociadas a cada acción. El data scientist deja al modelo elegir la secuencia de acciones que ejecutar para maximizar la recompensa total para cumplir un objetivo. 

Por lo tanto se encuentran a medio camino entre los modelos de aprendizaje supervisado y los modelos de aprendizaje no supervisado porque el data scientist le proporciona información sobre el escenario en el que va a trabajar de una forma similar a como se la proporciona a los modelos de aprendizaje supervisado, pero deja que el modelo escoja la mejor secuencia de acciones sin influir directamente, de forma similar a como trabajan los modelos de aprendizaje no supervisado. 

Otra diferencia importante es que los resultados de los modelos de aprendizaje por refuerzo dependen de los resultados previos que ha producido el modelo, dado que el objetivo es obtener la secuencia de acciones que nos lleve a maximizar la recompensa total. Desde el punto de vista matemático, podríamos describir la secuencia de acciones que escoge el modelo como una cadena de Markov.

# Conceptos básicos que nos van a acompañar en todos los modelos

Hay conceptos que son específicos de cada tipo de modelo y los iremos viendo en sus respectivas secciones, pero hay tres conceptos que son comunes a todos los modelos y que tenemos que tener en cuenta a la hora de crear nuestros modelos y de evaluarlos. 


## Varianza
El concepto de la varianza se refiere a la sensibilidad del modelo a los datos que utilicemos para el entrenamiento del modelo. 

Si cambiamos los datos de entrenamiento, podemos esperar una cierta varianza. Pero si hemos entrenado un modelo robusto, debería aprender las relaciones entre las variables de entrada y la variable de salida sin grandes cambios. 

Si el modelo presenta un alto valor de varianza, el modelo ha prestado mucha atención a los datos de entrenamiento y por lo tanto no generaliza bien para hacer predicciones sobre datos que no ha visto antes y produce grandes cambios en los resultados del modelo ante pequeños cambios en los datos de entrenamiento.


## Sesgo / Bias

El concepto de sesgo (bias) se refiere a la diferencia entre los valores predichos por el modelo y el valor real de los datos. 

Si el modelo presenta un alto valor de sesgo, entonces está prestando poca atención a los datos que le proporcionamos para entrenar y está simplificando en exceso el modelo. 

## Sobreajuste / Overfitting

Es el comportamiento que podemos ver en modelos que han aprendido el comportamiento de los datos de entrenamiento tan bien, que acaban teniendo problemas para generalizar y poder hacer buenas predicciones para datos que no ha visto nunca. Estos modelos tienen un bajo sesgo (bias) y una alta varianza.

## Subajuste / Underfitting

Es el comportamiento que podemos ver en modelos que no son capaces de aprender el comportamiento de los datos de entrenamiento. Estos modelos tienen un alto sesgo (bias) y una baja varianza. 


Tenemos que encontrar un equilibrio entre el sesgo (bias) y la varianza del modelo para que nuestro modelo pueda producir predicciones que sean buenas para todos los datos que reciba, los que recibe durante el entrenamiento y aquellos que recibe después y que no ha visto nunca. El underfitting normalmente se suele corregir añadiendo más datos al set de entrenamiento que le damos al modelo para que entrene.
