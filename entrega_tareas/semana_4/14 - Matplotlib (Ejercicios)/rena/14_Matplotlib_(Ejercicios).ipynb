````python
fig, axes = plt.subplots(1, 2, figsize=(6,4))
fig.tight_layout()
x = np.linspace(0,100, 1000)
# default grid appearance
axes[0].plot(x, 2*x, lw=2, linestyle='dashed', color = 'b')
axes[0].set_xlim([0,100])
axes[0].set_ylim([0,200])
axes[0].set_yticks(list(range(0,250,50)))
axes[0].tick_params(direction = 'in', rotation = 0)
# custom grid appearance
axes[1].plot(x, x**2, lw=4, color = 'r')
axes[1].set_xlim([0,100])
axes[1].set_ylim([0,10000])
axes[1].tick_params(direction = 'in', rotation = 0)
````

````python
# Ejes
ax = plt.axes(projection='3d')
# LÃ­nea
zline = np.linspace(0, 15, 1000)
xline = np.sin(zline)
yline = np.cos(zline)
# Nube de puntos
zdata = 15 * np.random.random(100)
xdata = np.sin(zdata) + 0.1 * np.random.randn(100)
ydata = np.cos(zdata) + 0.1 * np.random.randn(100)
#dibujar ahora la nube de puntos
ax.scatter3D(xdata, ydata, zdata, c=zdata, cmap='Greens')
#aÃ±adirle una lÃ­nea
ax.plot3D(xline, yline, zline, 'gray')
ax.view_init(20, 25)
````
