````python
# Creamos la clase
fig, ax = plt.subplots()
x = np.linspace(0, 100)

# par√°metros x, y, color
ax.plot(x, 2*x, 'b')

# Etiquetas de los ejes
ax.set_xlabel('x')
ax.set_ylabel('y')

# L√≠mites de los ejes, """"
ax.set_xlim([0,100])
ax.set_ylim([0,200])

# m√©todo con un par√°metro que mete los ticks dentro de la gr√°fica
ax.tick_params( direction= 'in')

# yticks = [0, 50, 100, 150, 200]
ax.set_yticks(list(range(0,250,50)))

# A√±adir titulo a la gr√°fica
ax.set_title('title')
````

````python
fig, ax = plt.subplots()
x = np.linspace(0, 100, 20)

# Gr√°fica principal 
ax.plot(x, 2*x, 'r')

# Nombre/Etiqueta de cada eje 
ax.set_xlabel('x')
ax.set_ylabel('y')

# L√≠mites de los ejes, el eje y no hace falta indicarlo xD
ax.set_xlim([0,100])
ax.set_ylim([0,200])

# yticks = [0, 50, 100, 150, 200]
ax.set_yticks(list(range(0,250,50)))

ax.tick_params(direction= 'in')

# inset, segunda gr√°fica "in" la gr√°fica que creamos al principio
inset_ax = fig.add_axes([0.28, 0.520, 0.20, 0.20]) # Posici√≥n= X, Y, Dimensiones= width, height
inset_ax.plot(x, 2*x, 'r')

# set axis range
inset_ax.set_xlim([0,100])
inset_ax.set_ylim([0,200])

# yticks = [0, 50, 100, 150, 200]
inset_ax.set_yticks(list(range(0,250,50)))
inset_ax.set_xticks(list(range(0,120,20)))

inset_ax.tick_params(direction= 'in')

ax.set_title('title')
````

````python
fig, ax = plt.subplots()
x = np.linspace(0, 100, 1000)

ax.plot(x, x**2, 'b')

ax.set_xlabel('X')
ax.set_ylabel('Z')

ax.set_xlim([0,100])
ax.set_ylim([0,200])

ax.tick_params(direction = 'in', rotation = 0)

ax.set_yticks(list(range(0,12000,2000)))
ax.set_title('title')

# inset, gr√°fico "in" 
inset_ax = fig.add_axes([0.28, 0.520, 0.31, 0.31]) # X, Y, width, height

# Titulo
inset_ax.set_title('zoom')

# set axis range
inset_ax.set_xlim([20, 22])
inset_ax.set_ylim([30, 50])

inset_ax.plot(x, x+20, 'b')
inset_ax.set_yticks(list(range(30,55,5)))
inset_ax.set_xticks(list(range(20,23,1)));
````

````python
fig, axes = plt.subplots(1, 2, figsize=(6,4))
fig.tight_layout()
x = np.linspace(0,100, 1000)

# default grid appearance
axes[0].plot(x, 2*x, lw=2, linestyle='dashed', color = 'b')
axes[0].set_xlim([0,100])
axes[0].set_ylim([0,200])
axes[0].set_yticks(list(range(0,250,50)))
axes[0].tick_params(direction = 'in', rotation = 0)

# custom grid appearance
axes[1].plot(x, x**2, lw=4, color = 'r')
axes[1].set_xlim([0,100])
axes[1].set_ylim([0,10000])
axes[1].tick_params(direction = 'in', rotation = 0)
````

````python
# Ejes
ax = plt.axes(projection='3d')

# L√≠nea
zline = np.linspace(0, 15, 1000)
xline = np.sin(zline)
yline = np.cos(zline)

# Nube de puntos
zdata = 15 * np.random.random(100)
xdata = np.sin(zdata) + 0.1 * np.random.randn(100)
ydata = np.cos(zdata) + 0.1 * np.random.randn(100)
 
# dibujar ahora la nube de puntos 
ax.scatter3D(xdata, ydata, zdata, c=zdata, cmap='Greens');

# a√±adirle una l√≠nea
ax.plot3D(xline, yline, zline, 'grey', lw=1)
````
