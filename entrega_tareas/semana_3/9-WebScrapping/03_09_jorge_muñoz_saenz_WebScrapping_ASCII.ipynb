```python
import requests

enlace = 'https://proyectoa.com/tabla-de-correspondencias-ascii-y-iso-8859-1/'

# Convertimos la web en texto
texto_web = requests.get(enlace).text
print(texto_web)
```

```python
# Le damos formato con Beautifulsoup
from bs4 import BeautifulSoup
sopa = BeautifulSoup(texto_web, 'lxml')
print(sopa)
```

```python
tablas_ascii = sopa.find_all('table', class_ ='')
print(tablas_ascii)
```

```python
import pandas as pd
dataframe_tablas_ascii = pd.read_html(str(tablas_ascii))[0]
print(dataframe_tablas_ascii)
```
