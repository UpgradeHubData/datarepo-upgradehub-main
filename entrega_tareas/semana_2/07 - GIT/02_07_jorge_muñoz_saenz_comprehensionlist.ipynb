# Ejercicio 4
# Elimina todas las vocales de una cadena (utiliza la cadena anterior)

```python
poema_sin_vocales = "".join([x for x in poema if x.upper() not in ["A", "E", "I", "O", "U"]])

print (poema_sin_vocales)
```

# Ejercicio 5
# Encontrar todas las palabras de una cadena que tengan menos de 5 letras (utilizar la cadena anterior)

```python
# Utilizamos el comando poema.split() para dividir la cadena en palabras
[palabra for palabra in poema.split() if (len(palabra) < 5)]
```

# Ejercicio 8
# Para todos los números del 1-1000, utilice una lista anidada/comprensión del diccionario para encontrar la cifra más alta por la que es divisible cualquiera de los números

```python
q8_answer = {
num:max([divisor for divisor in range(1,10) if num % divisor == 0]) for num in range(1001)
}

print (q8_answer)
```
