3. Contar el número de espacios en una cadena (utilizar la
cadena anterior)

```python
string = '''Región de manos sucias de pinceles sin pelo

de niños boca abajo de cepillos de dientes

Zona donde la rata se ennoblece

y hay banderas innúmeras y cantan himnos

y alguien te prende, hijo de puta

una medalla sobre el pecho

Y te pudres lo mismo'''
contador = 0
for x in string:
    if(x.isspace()):
        contador=contador+1
print("el numero de espacios es : ",contador)
```

4. Elimina todas las vocales de una cadena (utiliza la cadena
anterior)

```python
eliminavowels = "".join([x for x in string if x.upper() not in ["A", "E", "I", "O", "U"]])
print (eliminavowels)
```

5. Encontrar todas las palabras de una cadena que tengan
menos de 5 letras (utilizar la cadena anterior)

```python
palabras_menosde_5 = [x for x in string.split() if len(x)<5]
print (palabras_menosde_5)
```

6. Utilizar la comprensión de un diccionario para contar la
longitud de cada palabra de una frase (utilizar la cadena
anterior)

```python
contarlong = {
key:value for key, value in [(x, len(x)) for x in string.split()]
}
print (contarlong)
```
