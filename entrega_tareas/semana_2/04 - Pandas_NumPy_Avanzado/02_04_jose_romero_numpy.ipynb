##26 - Considere dos matrices aleatorias A y B, verifique si son iguales.

`pista: np.allclose, np.array_equal`

```python
A = np.random.randint(0,2,5)
B = np.random.randint(0,2,5)

# Si son identicas en forma pero no tienen los mismos valores (hay tolerancia).
equal = np.allclose(A,B)
print(equal)

# Si son identicas en forma y además tienen los mismos valores (no hay tolerancia).
equal = np.array_equal(A,B)
print(equal)
```

##29 - ¿Cómo encontrar el valor más cercano, a un escalar dado, en un vector?

`pista: argmin`


```python
Z = np.arange(100)
v = np.random.uniform(0,100)
index = (np.abs(Z-v)).argmin()
print(Z[index])
```

##30 - ¿Cómo convertir una matriz `float` (32 bits) en tipo `int` (32 bits)?

`pista: view and [:] =`

```python
Z = (np.random.rand(10)*100).astype(np.float32)
Y = Z.view(np.int32)
Y[:] = Z
print(Y)
```

##31 - ¿Cuál es el equivalente de `enumerate` para matrices en `numpy`?

`pista: np.ndenumerate, np.ndindex`

```python
Z = np.arange(16).reshape(4,4)
for index, value in np.ndenumerate(Z):
    print(index, value)
for index in np.ndindex(Z.shape):
    print(index, Z[index])
```

##33 - ¿Cómo ordenar una matriz por la enésima columna?

`pista: argsort`

```python
Z = np.random.randint(0,10,(3,3))
print(Z)
print("===========")
print(Z[Z[:,1].argsort()])
```

##34 - ¿Cómo saber si una matriz 2D tiene columnas nulas?

`pista: any, ~`


```python
Z = np.random.randint(0,3,(3,10))
print((~Z.any(axis=0)).any())
```

##35 - Encuentre el valor más cercano de un valor dado en una matriz.

`pista: np.abs, argmin, flat`


```python
Z = np.random.uniform(0,1,10)
z = 0.5
m = Z.flat[np.abs(Z - z).argmin()]
print(m)
```
